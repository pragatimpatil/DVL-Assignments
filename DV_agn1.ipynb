```python
#-----------------Question 1-----------------------
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

np.random.seed(42)

data = np.random.randn(100)
data = np.random.normal(0, 1, size=int(100))
# Matplotlib
plt.figure(figsize=(4, 3))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram - Matplotlib')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Seaborn
plt.figure(figsize=(4,3))
sns.histplot(data, bins=20, color='skyblue')
plt.title('Histogram - Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Plotly
fig = px.histogram(x=data, nbins=20, labels={'x': 'Value', 'y': 'Frequency'},
                   title='Histogram - Plotly')
fig.show()
```

```python
#----------------Question 2---------------
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# 90% from N(0, 1) and 10% from N(0, 25)
n1 = np.random.normal(0, 1, size=int(0.2 * 1000))
n2 = np.random.normal(0, 5, size=int(0.8 * 1000))

mixed = np.concatenate([n1, n2])

plt.hist(mixed, bins=75, density=False, alpha=0.8, color='red', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Mixture of Normals')

plt.show()
```

```python
#------------Question 3--------------
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, cauchy, probplot
np.random.seed(42)

n1 = np.random.normal(0, 1, size=int(0.9 * 1000))
n2 = np.random.normal(0, 5, size=int(0.1 * 1000))

mixed = np.concatenate([n1, n2])

# Create Q-Q plot with respect to quantiles of Cauchy(0, 1)
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
probplot(mixed, dist=cauchy(0, 1), plot=plt)
plt.title('Q-Q Plot - Cauchy(0, 1)-2nd sample')

# Create Q-Q plot with respect to quantiles of N(0, 1)
plt.subplot(2, 2, 2)
probplot(mixed, dist=norm(0, 1), plot=plt)
plt.title('Q-Q Plot - N(0, 1)-2nd sample')

plt.subplot(2, 2, 3)
probplot(data, dist=cauchy(0, 1), plot=plt)
plt.title('Q-Q Plot - Cauchy(0, 1)-1st sample')

plt.subplot(2, 2, 4)
probplot(data, dist=norm(0, 1), plot=plt)
plt.title('Q-Q Plot - N(0, 1)-1st sample')
plt.tight_layout()
plt.show()

#in the mixture of normals sample, the corner data points tend to be a lil away from the diagonal line wrt normal distribution
#as compared to the first sample data. The first normally sampled data almmost lie on the diagonal line and
# are closer to the normal distribution. There r more outliers in mixed sample wrt normal distribution-(plot[2,2,2]).

#in case of cauchy distribution, the mixed sample is somewhat closer to the diagonal line, hence their relative distance wrt to the line is lesser than that of the normally sampled data.
```

```python
#------------Question 4---------------
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

plt.figure(figsize=(8, 6))
plt.boxplot(mixed, vert=False)
plt.title('Box Plot of Second Sample')
plt.xlabel('Value')

q1 = np.percentile(mixed, 25)
q3 = np.percentile(mixed, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = [x for x in mixed if x < lower_bound or x > upper_bound]

# Print information about outliers and other statistics
print(f"Lower bound: {lower_bound}")
print(f"Upper bound: {upper_bound}")
print(f"Median: {np.median(mixed)}")
print(f"Mean: {np.mean(mixed)}")

plt.show()
#from the box plot, there are outliers beyond around -3 and +3. Median lies around 0
```
